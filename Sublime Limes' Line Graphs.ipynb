```python
from matplotlib import pyplot as plt
```

```python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

visits_per_month = [9695, 7909, 10831, 12942, 12495, 16794, 14161, 12762, 12777, 12439, 10309, 8724]

# numbers of limes of different species sold each month
key_limes_per_month = [92.0, 109.0, 124.0, 70.0, 101.0, 79.0, 106.0, 101.0, 103.0, 90.0, 102.0, 106.0]
persian_limes_per_month = [67.0, 51.0, 57.0, 54.0, 83.0, 90.0, 52.0, 63.0, 51.0, 44.0, 64.0, 78.0]
blood_limes_per_month = [75.0, 75.0, 76.0, 71.0, 74.0, 77.0, 69.0, 80.0, 63.0, 69.0, 73.0, 82.0]
```

```python
# Create a figure with the specified width and height
fig = plt.figure(figsize=(12, 8))

# Create the left subplot (1 row, 2 columns, subplot 1)
ax1 = fig.add_subplot(1, 2, 1)

x_values = range(len(months))

# Plot the total page visits
ax1.plot(x_values, visits_per_month, marker='o', color='b', linestyle='-')

# Add labels and title
ax1.set_xlabel('Months')
ax1.set_ylabel('Total Page Visits')
ax1.set_title('Total Page Visits over the Past Year')

# Set the x-axis ticks to be the months
ax1.set_xticks(x_values)

# Set the x-axis tick labels to be the names stored in the months list
ax1.set_xticklabels(months)

# Create the right subplot (1 row, 2 columns, subplot 2)
ax2 = fig.add_subplot(1, 2, 2)

# Plot the three lines representing lime sales against the months (x_values)
ax2.plot(months, key_limes_per_month, marker='o', color='r', linestyle='-', label='Key Limes')
ax2.plot(months, persian_limes_per_month, marker='s', color='g', linestyle='-', label='Persian Limes')
ax2.plot(months, blood_limes_per_month, marker='^', color='b', linestyle='-', label='Blood Limes')

# Add labels and title for the right subplot
ax2.set_xlabel('Months')
ax2.set_ylabel('Number of Limes Sold')
ax2.set_title('Lime Sales over the Past Year')
ax2.legend()  # Add a legend to differentiate the three lines

# Show the figure
plt.show()

# Save the figure as a PNG with a descriptive file name
filename = 'page_visits_and_lime_sales.png'
plt.savefig(filename)
```
